```python
df_authors_no=df_main.groupby('authors',as_index=False)[['title']].count().sort_values(by='title',ascending=False)[:15]
df_authors_no.rename(columns={'title':'No_of_books','authors':'Authors'},inplace=True)
plt.figure(figsize=(7,7))
sns.barplot(x='Authors',y='No_of_books',data=df_authors_no)
plt.xticks(rotation=90)
plt.title('No.of Books by an Author')
plt.show()
```

```python
df_authors_no
```

```python
df_title_pages=df_main.groupby(['title','authors'],as_index=False)[['# num_pages']].sum().sort_values('# num_pages',ascending=False)[:15]
plt.figure(figsize=(7,7))
sns.barplot(x='title',y='# num_pages',data=df_title_pages)
plt.xticks(rotation=90)
plt.title('No.of Pages in a Book')
plt.show()
```

```python
df_lang=df_main.groupby('language_code',as_index=False).count()[['language_code','title']].sort_values(by='title',ascending=False)[:15]
plt.figure(figsize=(7,7))
sns.barplot(x='language_code',y='title',data=df_lang)
plt.title('Languages VS No_of_books')
plt.show()
plt.hist(x='title',data=df_lang)
plt.title('Histogram on language code')
plt.show()
```
